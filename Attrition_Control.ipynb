```python
# Def_func to determine from both existing and left workers data,the departments with the highest percent of low salary earners
def low_salary_rate(low, medium, high):
    
    low_init = 0
    medium_init = 0
    high_init = 0
    for each_row in low.iterrows(): # looping through the rows and keeping a count of the low salary earners
        low_init += 1
    for each_row in medium.iterrows(): # looping through the rows and keeping a count of the medium salary earners
        medium_init += 1
    for each_row in high.iterrows(): # looping through the rows and keeping a count of the high salary earners
        high_init += 1
        
    total_salary_rows = low_init + medium_init + high_init
    rate_low_salary = low_init/total_salary_rows
    percent_low_salary = rate_low_salary * 100
    return round(percent_low_salary, 2)
```

```python
# Def-func to determine the promotion and work_accident rate for each of the departments for both the left and existing workers
def percent():
    
    try:
        num_of_dept = int(input("Enter number of department: ")) # number of department interested in checking
        init = num_of_dept # while loop initialization

        promotion_dept_dict_left = {}
        promotion_dept_dict_existing = {}
        work_accident_dept_dict_left = {}
        work_accident_dept_dict_existing = {}

        while init != 0:

            total_init_left = 0
            total_init_existing = 0
            ones_init_left = 0
            ones_init_existing = 0

            dept_name = input('Enter Dept name: ').strip()
            determinant = input('promotion or Work_accident: ').strip()
            
            if dept_name in list_of_dept_name and (determinant == 'promotion' or determinant == 'Work_accident'):

                dept_determinant_1_count_left = df_left_workers[(df_left_workers['dept'] == dept_name) & 
                                                                (df_left_workers[determinant] == 1)]
                dept_determinant_1_count_existing = df_existing_workers[(df_existing_workers['dept'] == dept_name) & 
                                                                        (df_existing_workers[determinant] == 1)]
                # df of number of workers with promotion/work_accident in the particular dept for both type of workers
                
                total_dept_row_count_left = df_left_workers[(df_left_workers['dept'] == dept_name) & 
                                                            ((df_left_workers[determinant] == 0)
                                                            | (df_left_workers[determinant] == 1))]
                total_dept_row_count_existing = df_existing_workers[(df_existing_workers['dept'] == dept_name) & 
                                                                    ((df_existing_workers[determinant] == 0)
                                                                    | (df_existing_workers[determinant] == 1))]
                # df of number of workers with & without promotion/work_accident in the particular dept for both dat of workers
                

                for i in dept_determinant_1_count_left.iterrows():
                    ones_init_left += 1
                for i in dept_determinant_1_count_existing.iterrows():
                    ones_init_existing += 1
                    
                # count of number of workers with promotion/work_accident in the particular dept for both type of workers

                for i in total_dept_row_count_left.iterrows():
                    total_init_left += 1
                for i in total_dept_row_count_existing.iterrows():
                    total_init_existing += 1
                    
                #count of num of workers with & without promotion/work_accident in the particular dept for both type of workers

                dept_percentage_left = round(((ones_init_left / total_init_left) * 100), 2)
                dept_percentage_existing = round(((ones_init_existing / total_init_existing) * 100), 2)

                if determinant == 'promotion':
                    promotion_dept_dict_left[dept_name] = dept_percentage_left
                    promotion_dept_dict_existing[dept_name] = dept_percentage_existing
                    
                else:
                    work_accident_dept_dict_left[dept_name] = dept_percentage_left
                    work_accident_dept_dict_existing[dept_name] = dept_percentage_existing

                init = init - 1
                
            else:
                print(f"{dept_name} or {determinant} does not exists.")
                
        sorted_promotion_dept_dict_left = sorted(promotion_dept_dict_left.items(), key=lambda x:x[1])
        sorted_promotion_dept_dict_existing = sorted(promotion_dept_dict_existing.items(), key=lambda x:x[1])
        
        sorted_work_accident_dept_dict_left = sorted(work_accident_dept_dict_left.items(), key=lambda x:x[1])
        sorted_work_accident_dept_dict_existing = sorted(work_accident_dept_dict_existing.items(), key=lambda x:x[1])
        
        # sorting the dict by values which automatically converts to a list
        
        sorted_promotion_dept_dict_convert_left = dict(sorted_promotion_dept_dict_left)
        sorted_promotion_dept_dict_convert_existing = dict(sorted_promotion_dept_dict_existing)
        
        sorted_work_accident_dept_dict_convert_left = dict(sorted_work_accident_dept_dict_left)
        sorted_work_accident_dept_dict_convert_existing = dict(sorted_work_accident_dept_dict_existing)
        
        # converting sorted list back to dict
        
        if len(sorted_promotion_dept_dict_convert_left) != 0 and len(sorted_promotion_dept_dict_convert_existing) != 0:
            print(f"Promotion percent of left workers: {sorted_promotion_dept_dict_convert_left}")
            print(f"Promotion percent of existing workers: {sorted_promotion_dept_dict_convert_existing}")
            
        if len(sorted_work_accident_dept_dict_convert_left) != 0 and len(sorted_work_accident_dept_dict_convert_existing) != 0:
            print(f"Work_accident percent of left workers: {sorted_work_accident_dept_dict_convert_left}")
            print(f"Work_accident percent of existing workers: {sorted_work_accident_dept_dict_convert_existing}")
            
    except Exception as err:
        print(f"Got {err} error")
        raise
```
